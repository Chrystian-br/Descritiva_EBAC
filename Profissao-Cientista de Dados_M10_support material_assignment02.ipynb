````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
sns.pointplot(x='posse_de_veiculo',y='renda',hue='posse_de_imovel',data=df)
````

````python
sns.pointplot(x='posse_de_imovel',y='renda',hue='posse_de_veiculo',data=df)
````

É possível observar o aumento da renda média entre os clientes que possuem ou não alguma posse, seja ela um imóvel ou um veículo (se for os dois, a renda é ainda maior).
Também fica claro que o erro padrão dos valores é baixo, indicando que existe uma evidência estatística de que a posse de um veículo ou imóvel interfere na renda média do cliente.

Desta forma, pode-se dizer que as variáveis explicativas qualitativas **posse de veiculo** e **posse de imovel** são boas preditoras para a variável resposta **renda**.
