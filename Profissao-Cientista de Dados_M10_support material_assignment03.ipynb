````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('previsao_de_renda.csv').drop(['Unnamed: 0'], axis=1)
df.head()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
sns.countplot(x='data_ref', hue='educacao', data=df)

plt.xticks(rotation=60);
````

Na variável qualitativa de **educação**, é possível perceber uma mudança grande na escolaridade dos clientes a partir de 01-08-2015, o que pode significar que, no geral, a escolaridade média dos clientes a partir desta data foi alterada e, possivelmente, a renda.
Diferença esta que não é muito perceptível ao repetir o processo com a variável **posse de veiculo**:

````python
sns.countplot(x='data_ref', hue='posse_de_veiculo', data=df)

plt.xticks(rotation=60);
````

2) Para cada variável quantitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
df.select_dtypes(['int','float']).head()
````

````python
sns.pointplot(x='data_ref', y='qtd_filhos', data=df)
plt.xticks(rotation=60);
````

Como pode se perceber, a variável referente à quantidade de filhos não possui alguma alteração preocupante durante o tempo, mantendo-se entre 0.40 e 0.45 por grande parte do ano, tendo apenas um pico que alcança o valor de 0.50 com um erro padrão alcançando até 0.55. Sendo asssim, pode-se dizer que a variável possui uma boa estabilidade e mantém um modelo preciso.

---------

````python
sns.pointplot(x='data_ref', y='idade', data=df)
plt.xticks(rotation=60);
````

A variável **idade** mantém uma estabilidade muito semelhante à variável **qtd_filhos**, oscilando com diferenças pequenas e insuficientes para causar dúvidas sobre a precisão de uma modelagem.

---------

````python
sns.pointplot(x='data_ref', y='tempo_emprego', data=df)
plt.xticks(rotation=60);
````

A variável referente ao tempo de emprego do cliente possui um erro padrão um pouco alto em alguns pontos, alcançando uma diferença média de quase 1 ano em alguns meses, podendo mostrar certa imprecisão ou instabilidade no momento de previsão da modelagem.

---------

````python
sns.pointplot(x='data_ref', y='qt_pessoas_residencia', data=df)
plt.xticks(rotation=60);
````

**qt_pessoas_residencia** parece ser uma variável boa para ser utilizada em uma modelagem, dado ao fato de, apesar de seu gráfico possuir algumas oscilações grandes, a diferença dos valores médios é muito baixa e possui certa estabilidade.
